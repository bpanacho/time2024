```python
import json
import os

def convert_json_to_yolo(json_folder, output_folder, viewport_width=1920, viewport_height=1080):
    # Verificar se a pasta de saída existe, caso contrário, criar
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterar sobre cada arquivo JSON
    for json_file in os.listdir(json_folder):
        if json_file.endswith(".json"):
            with open(os.path.join(json_folder, json_file)) as f:
                data = json.load(f)

            frame_index = data["frame_index"]
            people = data["2dPlayerTracking"]["people"]

            # Criar um arquivo .txt para cada frame
            output_file_path = os.path.join(output_folder, f"{frame_index}.txt")
            with open(output_file_path, "w") as txt_file:
                for person in people:
                    # Extrair coordenadas da caixa delimitadora
                    left = person["left"]
                    top = person["top"]
                    right = person["right"]
                    bottom = person["bottom"]

                    # Calcular x_center, y_center, width, height normalizados
                    x_center = ((left + right) / 2) / viewport_width
                    y_center = ((top + bottom) / 2) / viewport_height
                    width = (right - left) / viewport_width
                    height = (bottom - top) / viewport_height

                    # Escrever no formato YOLO: class_id x_center y_center width height
                    # Assumindo que "0" é a classe do lutador
                    txt_file.write(f"0 {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
            
            print(f"Arquivo YOLO salvo em: {output_file_path}")

# Exemplo de uso
json_folder = "D:/hackathon/boxe/2024_07_13_Albert Gonzalez_Conrado S. Martinez/targets"
output_folder = "/home/bianca-panacho/Desktop/HACKA/boxe/dataset/train/labels"
convert_json_to_yolo(json_folder, output_folder
```

```python
import os
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt

def carregar_imagem(caminho):
    """Carrega a imagem e converte para HSV."""
    img = cv2.imread(caminho)
    if img is None:
        raise FileNotFoundError(f"Erro ao carregar a imagem: {caminho}")
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    return img, hsv

def aplicar_mascara_verde(hsv, lower_green=(30, 40, 40), upper_green=(80, 255, 255)):
    """Aplica uma máscara para isolar a área verde na imagem HSV."""
    mask = cv2.inRange(hsv, np.array(lower_green), np.array(upper_green))
    mask = cv2.GaussianBlur(mask, (5, 5), 0)  # Suavizar a máscara para reduzir ruídos
    return mask

def aplicar_mascara_azul(hsv, lower_blue=(100, 150, 0), upper_blue=(140, 255, 255)):
    """Aplica uma máscara para isolar a área azul do chão do ringue e excluí-la."""
    mask = cv2.inRange(hsv, np.array(lower_blue), np.array(upper_blue))
    mask = cv2.GaussianBlur(mask, (5, 5), 0)
    return mask

def calcular_cor_media(img, mask):
    """Calcula a cor média na área verde da imagem."""
    media_cor = cv2.mean(img, mask=mask)[:3]  # Apenas valores BGR
    return media_cor

def processar_imagens(pasta_imagens, pasta_saida, n_exibir=3):
    """Processa todas as imagens na pasta, aplica a máscara verde e salva as imagens."""
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    imagens_disponiveis = [img for img in os.listdir(pasta_imagens) if img.endswith('.jpg') or img.endswith('.jpeg')]
    imagens_selecionadas = random.sample(imagens_disponiveis, min(n_exibir, len(imagens_disponiveis)))
    
    # Lista para armazenar imagens para exibição
    imagens_para_exibir = []

    for nome_imagem in imagens_disponiveis:
        caminho_imagem = os.path.join(pasta_imagens, nome_imagem)
        img, hsv = carregar_imagem(caminho_imagem)

        # Aplicar a máscara para a cor verde
        mask_verde = aplicar_mascara_verde(hsv)

        # Aplicar a máscara para a cor azul (chão do ringue)
        mask_azul = aplicar_mascara_azul(hsv)

        # Remover o chão do ringue da máscara verde
        mask_verde = cv2.bitwise_and(mask_verde, cv2.bitwise_not(mask_azul))
        
        # Aplicar a máscara resultante à imagem original para exibir apenas a área verde sem o chão azul
        masked_img = cv2.bitwise_and(img, img, mask=mask_verde)

        # Salvar a imagem com a máscara aplicada
        caminho_saida = os.path.join(pasta_saida, nome_imagem)
        cv2.imwrite(caminho_saida, masked_img)
        
        # Calcular a cor média da área verde (opcional)
        grass_color = calcular_cor_media(img, mask_verde)
        print(f"Cor média da área verde em {nome_imagem} (BGR):", grass_color)
        
        # Adiciona à lista para exibir se estiver entre as selecionadas
        if nome_imagem in imagens_selecionadas:
            imagens_para_exibir.append(masked_img)
            print(f"Imagem adicionada para exibição: {nome_imagem}")

    # Verificar se temos imagens para exibir
    if imagens_para_exibir:
        plt.figure(figsize=(15, 5))
        for i, img in enumerate(imagens_para_exibir):
            plt.subplot(1, n_exibir, i + 1)
            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            plt.axis("off")
            plt.title(f"Exemplo {i + 1}")
        plt.show()
    else:
        print("Nenhuma imagem foi selecionada para exibição.")

# Caminhos de entrada e saída
pasta_imagens = 'D:/Projeto/hackfot/final-boxe/dataset/val/images'
pasta_saida = 'D:/Projeto/hackfot/final-boxe/output'

# Processar as imagens e exibir três imagens aleatórias
processar_imagens(pasta_imagens, pasta_saida, n_exibir=3)
```

```python
import os
import json
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt

def carregar_imagem(caminho):
    """Carrega a imagem e converte para HSV."""
    img = cv2.imread(caminho)
    if img is None:
        raise FileNotFoundError(f"Erro ao carregar a imagem: {caminho}")
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    return img, hsv

def carregar_labels(caminho_json):
    """Carrega as coordenadas das box labels dos lutadores de um arquivo JSON."""
    with open(caminho_json, 'r') as f:
        dados = json.load(f)
    boxes = []
    for pessoa in dados.get("2dPlayerTracking", {}).get("people", []):
        # Converte as coordenadas para inteiros
        left = int(pessoa['left'])
        top = int(pessoa['top'])
        right = int(pessoa['right'])
        bottom = int(pessoa['bottom'])
        boxes.append((left, top, right, bottom))
    return boxes

def aplicar_mascara_lutadores(img, boxes):
    """Aplica uma máscara que mantém apenas as áreas dentro das caixas delimitadoras (boxes) dos lutadores."""
    # Cria uma máscara preta (tudo fora das boxes será preto)
    mask = np.zeros(img.shape[:2], dtype=np.uint8)
    
    # Preenche a máscara com branco nas áreas das boxes dos lutadores
    for (left, top, right, bottom) in boxes:
        cv2.rectangle(mask, (left, top), (right, bottom), 255, -1)
    
    # Aplica a máscara para manter apenas as áreas das boxes visíveis na imagem original
    lutadores_img = cv2.bitwise_and(img, img, mask=mask)
    return lutadores_img

def processar_imagens_lutadores(pasta_imagens, pasta_json, pasta_saida, n_exibir=3):
    """Processa todas as imagens na pasta, aplica a máscara para isolar os lutadores e salva as imagens."""
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    imagens_disponiveis = [img for img in os.listdir(pasta_imagens) if img.endswith('.jpg') or img.endswith('.jpeg')]
    imagens_selecionadas = random.sample(imagens_disponiveis, min(n_exibir, len(imagens_disponiveis)))
    
    for nome_imagem in imagens_disponiveis:
        caminho_imagem = os.path.join(pasta_imagens, nome_imagem)
        img, hsv = carregar_imagem(caminho_imagem)
        
        # Nome do JSON correspondente ao frame
        nome_json = nome_imagem.replace('.jpg', '.json').replace('.jpeg', '.json')
        caminho_json = os.path.join(pasta_json, nome_json)
        
        # Carrega as coordenadas das boxes dos lutadores
        if os.path.exists(caminho_json):
            boxes = carregar_labels(caminho_json)
            # Aplicar a máscara para manter apenas as áreas dos lutadores
            lutadores_img = aplicar_mascara_lutadores(img, boxes)
            
            # Salvar a imagem com os lutadores isolados
            caminho_saida = os.path.join(pasta_saida, nome_imagem)
            cv2.imwrite(caminho_saida, lutadores_img)
            print(f"Imagem salva com lutadores isolados: {nome_imagem}")
    
    # Exibir 3 imagens aleatórias com apenas os lutadores e fundo preto
    plt.figure(figsize=(15, 5))
    for i, nome_imagem in enumerate(imagens_selecionadas):
        caminho_saida = os.path.join(pasta_saida, nome_imagem)
        img = cv2.imread(caminho_saida)
        
        plt.subplot(1, n_exibir, i + 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.title(nome_imagem)
    
    plt.show()

# Caminhos de entrada e saída
pasta_imagens = 'D:/Projeto/hackfot/final-boxe/dataset/val/images'
pasta_json = 'D:/Projeto/hackfot/final-boxe/dataset/val/labels'
pasta_saida = 'D:/Projeto/hackfot/final-boxe/output'

# Processar as imagens e exibir três imagens aleatórias com lutadores isolados
processar_imagens_lutadores(pasta_imagens, pasta_json, pasta_saida, n_exibir=3)
```

```python
import os
import json
import cv2
import numpy as np
import matplotlib.pyplot as plt

def carregar_imagem(caminho):
    """Carrega a imagem."""
    img = cv2.imread(caminho)
    if img is None:
        raise FileNotFoundError(f"Erro ao carregar a imagem: {caminho}")
    return img

def carregar_labels(caminho_json):
    """Carrega as coordenadas das box labels dos lutadores de um arquivo JSON."""
    with open(caminho_json, 'r') as f:
        dados = json.load(f)
    boxes = []
    for pessoa in dados.get("2dPlayerTracking", {}).get("people", []):
        # Converte as coordenadas para inteiros
        left = int(pessoa['left'])
        top = int(pessoa['top'])
        right = int(pessoa['right'])
        bottom = int(pessoa['bottom'])
        boxes.append((left, top, right, bottom))
    return boxes

def refinar_segmentacao(img, box):
    """
    Refina a segmentação aplicando limiarização e operações morfológicas
    para isolar os lutadores dentro da área da box.
    """
    left, top, right, bottom = box
    roi = img[top:bottom, left:right]  # Extrai a região de interesse (ROI) da box
    
    # Converter a ROI para escala de cinza
    gray_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    
    # Aplicar limiar adaptativo para destacar o lutador
    _, thresh_roi = cv2.threshold(gray_roi, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Operações morfológicas para refinar a máscara do lutador
    kernel = np.ones((3, 3), np.uint8)
    morphed_roi = cv2.morphologyEx(thresh_roi, cv2.MORPH_CLOSE, kernel)
    morphed_roi = cv2.morphologyEx(morphed_roi, cv2.MORPH_OPEN, kernel)
    
    # Aplica a máscara resultante à ROI para isolar o lutador
    segmentado_roi = cv2.bitwise_and(roi, roi, mask=morphed_roi)
    
    # Substitui a ROI segmentada na imagem original
    img[top:bottom, left:right] = segmentado_roi
    return img

def processar_imagens_segmentadas(pasta_imagens, pasta_json, pasta_saida, n_exibir=3):
    """Processa todas as imagens na pasta, aplica segmentação refinada dentro das boxes e salva as imagens."""
    if not os.path.exists(pasta_saida):
        os.makedirs(pasta_saida)

    imagens_disponiveis = [img for img in os.listdir(pasta_imagens) if img.endswith('.jpg') or img.endswith('.jpeg')]
    imagens_selecionadas = random.sample(imagens_disponiveis, min(n_exibir, len(imagens_disponiveis)))
    
    for nome_imagem in imagens_disponiveis:
        caminho_imagem = os.path.join(pasta_imagens, nome_imagem)
        img = carregar_imagem(caminho_imagem)
        
        # Nome do JSON correspondente ao frame
        nome_json = nome_imagem.replace('.jpg', '.json').replace('.jpeg', '.json')
        caminho_json = os.path.join(pasta_json, nome_json)
        
        # Carrega as coordenadas das boxes dos lutadores
        if os.path.exists(caminho_json):
            boxes = carregar_labels(caminho_json)
            # Refina a segmentação para cada box de lutador
            for box in boxes:
                img = refinar_segmentacao(img, box)
            
            # Salvar a imagem com os lutadores segmentados e fundo preto
            caminho_saida = os.path.join(pasta_saida, nome_imagem)
            cv2.imwrite(caminho_saida, img)
            print(f"Imagem salva com lutadores segmentados dentro das boxes: {nome_imagem}")
    
    # Exibir 3 imagens aleatórias com os lutadores segmentados e fundo preto dentro das boxes
    plt.figure(figsize=(15, 5))
    for i, nome_imagem in enumerate(imagens_selecionadas):
        caminho_saida = os.path.join(pasta_saida, nome_imagem)
        img = cv2.imread(caminho_saida)
        
        plt.subplot(1, n_exibir, i + 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.title(nome_imagem)
    
    plt.show()

# Caminhos de entrada e saída
pasta_imagens = 'D:/Projeto/hackfot/final-boxe/output'  # Saída do código anterior
pasta_json = 'D:/Projeto/hackfot/final-boxe/dataset/val/labels'
pasta_saida = 'D:/Projeto/hackfot/final-boxe/output_segmentado_preciso'

# Processar as imagens e exibir três imagens aleatórias com lutadores segmentados dentro das boxes
processar_imagens_segmentadas(pasta_imagens, pasta_json, pasta_saida, n_exibir=3)
```

```python
import os
import json
import random
import numpy as np
import cv2
import matplotlib.pyplot as plt
from collections import Counter

def carregar_jsons(pasta_json):
    """
    Carrega todos os arquivos JSON de uma pasta e retorna um dicionário
    com o nome do arquivo como chave e o conteúdo JSON como valor.
    """
    dados = {}
    for arquivo in os.listdir(pasta_json):
        if arquivo.endswith('.json'):
            caminho_arquivo = os.path.join(pasta_json, arquivo)
            with open(caminho_arquivo, 'r') as f:
                dados[arquivo] = json.load(f)
    return dados

def cor_predominante(img, mask):
    """Calcula a cor mais frequente na área selecionada da imagem, desconsiderando preto."""
    # Aplica a máscara na imagem
    masked_img = cv2.bitwise_and(img, img, mask=mask)
    
    # Reshape para uma lista de pixels
    pixels = masked_img.reshape(-1, 3)
    
    # Remove pixels próximos do preto
    pixels = [tuple(pixel) for pixel in pixels if np.any(pixel > 30)]  # Threshold para ignorar preto
    
    if pixels:
        # Conta as ocorrências de cada cor
        frequencias = Counter(pixels)
        # Obtém a cor mais comum
        cor_mais_comum = frequencias.most_common(1)[0][0]
        return np.array(cor_mais_comum)
    else:
        return np.array([0, 0, 0])  # Retorna preto se não houver outra cor dominante

def exibir_resultados(img, player_kit_color):
    """
    Exibe a cor predominante do uniforme do jogador e o corte da imagem correspondente.
    """
    # Exibir a imagem do jogador
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()

    # Exibir a cor predominante do uniforme
    demo_color = np.ones((50, 50, 3), dtype=np.uint8) * player_kit_color.astype(np.uint8)
    plt.imshow(cv2.cvtColor(demo_color, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()

    print("Player Kit Predominant Color: ", player_kit_color)
    print('#' * 100)

def processar_parte_superior(player_img, color_range):
    """Corta a imagem para manter apenas a parte superior e aplica a segmentação da cor especificada."""
    altura = player_img.shape[0]
    parte_superior = player_img[:altura // 2, :]

    # Converter para HSV
    hsv = cv2.cvtColor(parte_superior, cv2.COLOR_BGR2HSV)
    
    # Aplicar máscara de segmentação com a faixa de cor definida
    mask = cv2.inRange(hsv, color_range[0], color_range[1])
    
    # Extrair apenas a cor da máscara aplicada
    img_segmentada = cv2.bitwise_and(parte_superior, parte_superior, mask=mask)

    return img_segmentada, mask

# Definições de faixa de cor para o lutador de shorts vermelhos e outro lutador
red_range = (np.array([0, 120, 70]), np.array([10, 255, 255]))  # Faixa de vermelho
other_range = (np.array([90, 50, 50]), np.array([130, 255, 255]))  # Exemplo de faixa azul

# Caminhos das pastas
pasta_imagens = 'D:/Projeto/hackfot/final-boxe/dataset/val/images'
pasta_json = 'D:/Projeto/hackfot/final-boxe/dataset/val/labels'

# Carrega os dados JSON
dados_json = carregar_jsons(pasta_json)

# Dicionário para armazenar as cores predominantes de cada jogador
cores_jogadores = {}

# Seleciona todas as imagens disponíveis para processar
imagens_disponiveis = [img for img in os.listdir(pasta_imagens) if img.endswith('.jpg') or img.endswith('.png')]

# Processa todas as imagens da pasta
for nome_imagem in imagens_disponiveis:
    caminho_imagem = os.path.join(pasta_imagens, nome_imagem)
    img = cv2.imread(caminho_imagem)

    nome_json = os.path.splitext(nome_imagem)[0] + '.json'
    if nome_json in dados_json:
        deteccoes = dados_json[nome_json]
        
        cores_jogadores[nome_imagem] = []  # Inicializa uma lista para essa imagem

        for pessoa in deteccoes.get("2dPlayerTracking", {}).get("people", []):
            x1, y1, x2, y2 = int(pessoa['left']), int(pessoa['top']), int(pessoa['right']), int(pessoa['bottom'])
            player_img = img[y1: y2, x1: x2]

            # Processar a parte superior do jogador para diferentes faixas de cor
            camisa_img_red, mask_red = processar_parte_superior(player_img, red_range)
            camisa_img_other, mask_other = processar_parte_superior(player_img, other_range)

            # Calcular a cor predominante para o vermelho e a outra cor
            player_kit_color_red = cor_predominante(camisa_img_red, mask_red)
            player_kit_color_other = cor_predominante(camisa_img_other, mask_other)

            # Salva as cores predominantes no dicionário
            cores_jogadores[nome_imagem].append({
                "red_kit_color": player_kit_color_red.tolist(),
                "other_kit_color": player_kit_color_other.tolist()
            })

# Seleciona três imagens aleatórias para exibir os resultados
imagens_para_exibir = random.sample(imagens_disponiveis, min(3, len(imagens_disponiveis)))

for nome_imagem in imagens_para_exibir:
    caminho_imagem = os.path.join(pasta_imagens, nome_imagem)
    img = cv2.imread(caminho_imagem)

    nome_json = os.path.splitext(nome_imagem)[0] + '.json'
    if nome_json in dados_json:
        deteccoes = dados_json[nome_json]

        for pessoa in deteccoes.get("2dPlayerTracking", {}).get("people", []):
            x1, y1, x2, y2 = int(pessoa['left']), int(pessoa['top']), int(pessoa['right']), int(pessoa['bottom'])
            player_img = img[y1: y2, x1: x2]

            camisa_img_red, mask_red = processar_parte_superior(player_img, red_range)
            camisa_img_other, mask_other = processar_parte_superior(player_img, other_range)

            player_kit_color_red = cor_predominante(camisa_img_red, mask_red)
            player_kit_color_other = cor_predominante(camisa_img_other, mask_other)

            # Exibir os resultados para análise
            exibir_resultados(camisa_img_red, player_kit_color_red)
            exibir_resultados(camisa_img_other, player_kit_color_other)

# Salvar as cores em um arquivo JSON
caminho_cores_json = 'D:/Projeto/hackfot/final-boxe/output_segmentado_preciso/cores-jogadores.json'
with open(caminho_cores_json, 'w') as f:
    json.dump(cores_jogadores, f, indent=4)

print(f"Cores salvas em {caminho_cores_json}")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

def exibir_kits_colors(kits_colors):
    """
    Exibe uma imagem com as cores dos uniformes processadas a partir do array fornecido.
    Cada quadrado na imagem representa a cor média de um uniforme.

    Parâmetros:
    - kits_colors: array de cores médias dos uniformes (RGB).
    """
    # Número total de cores a serem exibidas
    num_colors = len(kits_colors)
    
    # Calcula o tamanho do grid para exibir todas as cores (aproximadamente quadrado)
    grid_rows = int(np.ceil(np.sqrt(num_colors)))  # Número de linhas
    grid_cols = grid_rows  # Número de colunas será igual ao de linhas para formar um grid quadrado

    # Cria uma imagem com espaço para todos os quadrados de cores
    color_image = np.zeros((grid_rows * 50, grid_cols * 50, 3), dtype=np.uint8)

    # Preenche a imagem com as cores do array
    for i, color in enumerate(kits_colors):
        row = i // grid_cols
        col = i % grid_cols
        color_image[row*50:(row+1)*50, col*50:(col+1)*50] = color  # Preenche quadrados 50x50 pixels

    # Ajusta o tamanho da imagem para que as cores fiquem em destaque
    plt.figure(figsize=(grid_cols, grid_rows))
    plt.imshow(color_image)
    plt.axis("off")
    plt.title("Cores dos Uniformes Processadas")
    plt.show()

def ler_cores_do_json(caminho_json):
    """
    Lê as cores médias dos uniformes a partir de um arquivo JSON.
    
    Parâmetros:
    - caminho_json: caminho do arquivo JSON.
    
    Retorna:
    - Um array de cores médias (RGB).
    """
    with open(caminho_json, 'r') as f:
        dados = json.load(f)
    
    # Criar uma lista para armazenar todas as cores
    todas_cores = []
    
    # Itera por todas as imagens e cores no arquivo JSON
    for imagem, cores in dados.items():
        # Cada cor precisa ser convertida de float para int e adicionada à lista
        for cor_obj in cores:
            if "red_kit_color" in cor_obj:
                red_kit_color = [int(c) for c in cor_obj["red_kit_color"]]
                todas_cores.append(red_kit_color)
            if "other_kit_color" in cor_obj:
                other_kit_color = [int(c) for c in cor_obj["other_kit_color"]]
                todas_cores.append(other_kit_color)
    
    return todas_cores

# Caminho do arquivo JSON
caminho_json = 'D:/Projeto/hackfot/final-boxe/output_segmentado_preciso/cores-jogadores.json'

# Lê as cores do arquivo JSON
kits_colors = ler_cores_do_json(caminho_json)

# Exibe as cores processadas
exibir_kits_colors(kits_colors)
```

```python
import cv2
import numpy as np
import os
import json
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import random

# Carregar cores dos uniformes do JSON
json_path = "D:/Projeto/hackfot/pqpdeubom/cores_jogadores/cores-jogadores.json"
with open(json_path, 'r') as f:
    data = json.load(f)

# Extrair cores
kits_colors = []
for image_colors in data.values():
    kits_colors.extend(image_colors)  
kits_colors = np.array(kits_colors)

# Treinar o modelo KMeans para agrupar cores similares
kmeans = KMeans(n_clusters=2, random_state=3, n_init=10).fit(kits_colors)
centers = kmeans.cluster_centers_

# Caminho para pastas
pasta_imagens = 'D:/Projeto/hackfot/final-boxe/dataset/train/images'
pasta_json = 'D:/Projeto/hackfot/final-boxe/dataset/train/labels'
pasta_output = 'D:/Projeto/hackfot/final-boxe/output'

# Definir a cor aproximada do uniforme do juiz
cor_juiz = (0, 0, 0)  # Ajuste a cor para a média aproximada do uniforme do juiz
intensidade_limite = 50  # Limite de intensidade para considerar uma cor "escura" (ajuste conforme necessário)

# Dicionário para armazenar a correspondência de cores e classes
memoria_classes = {}

# Função para extrair a cor média
def calcular_cor_media(img, x1, y1, x2, y2):
    roi = img[y1:y2, x1:x2]
    media_cor = np.mean(roi, axis=(0, 1))  
    return tuple(media_cor)

# Processar e classificar imagem
def processar_e_classificar_imagem(img_path, json_path, kmeans, centers, cor_juiz, memoria_classes):
    img = cv2.imread(img_path)
    with open(json_path, 'r') as f:
        dados_json = json.load(f)

    # Lista para rastrear classes já atribuídas
    classes_atribuidas = []

    for pessoa in dados_json.get("2dPlayerTracking", {}).get("people", []):
        x1, y1, x2, y2 = int(pessoa['left']), int(pessoa['top']), int(pessoa['right']), int(pessoa['bottom'])
        
        media_cor = calcular_cor_media(img, x1, y1, x2, y2)
        
        # Verificar se a cor média está próxima da cor do juiz
        if np.allclose(media_cor, cor_juiz, atol=30):  # Tolerância de 30 para ignorar variações
            continue  # Ignora o juiz e passa para a próxima detecção

        # Verificar se a cor já foi classificada anteriormente
        if media_cor in memoria_classes:
            cluster_label = memoria_classes[media_cor]  # Recupera a classe da memória
        else:
            # Priorizar "Time 0" para cores escuras
            intensidade_media = np.mean(media_cor)
            if intensidade_media < intensidade_limite and 0 not in classes_atribuidas:
                cluster_label = 0  # Define como "Time 0" para cores escuras
            else:
                # Usar KMeans para classificar jogadores com base na cor média
                media_cor_reshape = np.array(media_cor).reshape(1, -1)
                predicted_label = kmeans.predict(media_cor_reshape)[0]
                
                # Escolher a classe mais próxima que ainda não foi atribuída
                if predicted_label in classes_atribuidas:
                    distancias = np.linalg.norm(centers - np.array(media_cor), axis=1)
                    possiveis_labels = [i for i in range(len(centers)) if i not in classes_atribuidas]
                    cluster_label = possiveis_labels[np.argmin(distancias[possiveis_labels])]
                else:
                    cluster_label = predicted_label
            
            # Salvar a nova correspondência de cor e classe na memória
            memoria_classes[media_cor] = cluster_label

        # Marcar a classe como atribuída
        classes_atribuidas.append(cluster_label)

        # Definir cor da caixa para cada time
        if cluster_label == 0:
            cor_box = (255, 0, 0)  # Vermelho para Time 0
        elif cluster_label == 1:
            cor_box = (0, 0, 255)  # Azul para Time 1
        else:
            cor_box = (0, 255, 0)  # Verde para Time 2

        # Desenhar caixa e rótulo
        cv2.rectangle(img, (x1, y1), (x2, y2), cor_box, 2)
        cv2.putText(img, f'Time {cluster_label}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, cor_box, 2)

    return img

# Selecionar imagens para teste
imagens = [img for img in os.listdir(pasta_imagens) if img.endswith((".jpg", ".png"))]
for nome_imagem in imagens:
    caminho_imagem = os.path.join(pasta_imagens, nome_imagem)
    caminho_json = os.path.join(pasta_json, os.path.splitext(nome_imagem)[0] + '.json')

    if os.path.exists(caminho_json):
        img_processada = processar_e_classificar_imagem(caminho_imagem, caminho_json, kmeans, centers, cor_juiz, memoria_classes)
        caminho_saida = os.path.join(pasta_output, nome_imagem)
        cv2.imwrite(caminho_saida, img_processada)

# Selecionar três imagens aleatórias para exibição
imagens_processadas = random.sample(imagens, min(3, len(imagens)))
for i, nome_imagem in enumerate(imagens_processadas):
    caminho_saida = os.path.join(pasta_output, nome_imagem)
    img = cv2.imread(caminho_saida)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6, 6))
    plt.imshow(img_rgb)
    plt.title(f"Imagem {i + 1}")
    plt.axis("off")
plt.show()
```

```python
import json
import os

# Dimensões da imagem (ajuste de acordo com sua imagem)
imagem_largura = 1280  # Largura da imagem
imagem_altura = 720    # Altura da imagem

# Caminho para a pasta de entrada e saída
pasta_json_entrada = 'D:/Projeto/hackfot/pqpdeubom/classificacoes_finais'
pasta_saida = 'D:/Projeto/hackfot/pqpdeubom/classificacoes_convertidas'
os.makedirs(pasta_saida, exist_ok=True)

# Função para converter os dados do JSON para o formato desejado
def converter_para_formato_desejado(json_path, output_path, largura_img, altura_img):
    with open(json_path, 'r') as f:
        dados = json.load(f)
    
    linhas_saida = []
    for pessoa in dados.get("2dPlayerTracking", {}).get("people", []):
        # Obter coordenadas e time
        left = pessoa['left']
        top = pessoa['top']
        right = pessoa['right']
        bottom = pessoa['bottom']
        time = pessoa['team']
        
        # Converter o time para a classe (0 ou 1)
        classe = 0 if time == "Time 0" else 1

        # Calcular coordenadas normalizadas
        center_x = (left + right) / 2 / largura_img
        center_y = (top + bottom) / 2 / altura_img
        width = (right - left) / largura_img
        height = (bottom - top) / altura_img

        # Adicionar a linha no formato desejado
        linha = f"{classe} {center_x} {center_y} {width} {height}"
        linhas_saida.append(linha)
    
    # Salvar no arquivo de saída
    with open(output_path, 'w') as f:
        f.write("\n".join(linhas_saida))

# Processar todos os arquivos JSON na pasta de entrada
for nome_arquivo in os.listdir(pasta_json_entrada):
    if nome_arquivo.endswith(".json"):
        caminho_json = os.path.join(pasta_json_entrada, nome_arquivo)
        caminho_saida = os.path.join(pasta_saida, os.path.splitext(nome_arquivo)[0] + '.txt')
        
        # Converter e salvar no formato desejado
        converter_para_formato_desejado(caminho_json, caminho_saida, imagem_largura, imagem_altura)
```

```python
import cv2
import os
import json
import numpy as np

# Caminhos para as pastas de entrada e saída
pasta_imagens = 'D:/Projeto/hackfot/final-boxe/dataset/train/images/lutador-classificado'  # Local das imagens classificadas
pasta_json = 'D:/Projeto/hackfot/final-boxe/dataset/train/labels'  # Local dos arquivos JSON de entrada
pasta_output = 'D:/Projeto/hackfot/final-boxe/dataset/train/saidas-final-json'  # Local de saída dos novos arquivos JSON com classificação

# Cria a pasta de saída, se não existir
os.makedirs(pasta_output, exist_ok=True)

# Função para classificar lutadores com base na posição
def classificar_lutadores(img_path, json_path, output_json_path):
    # Carregar a imagem e o JSON de coordenadas
    img = cv2.imread(img_path)
    with open(json_path, 'r') as f:
        dados_json = json.load(f)
    
    # Estrutura para o novo JSON com as classificações
    classificacoes = {
        "frame_index": dados_json.get("frame_index"),
        "viewport_width": dados_json.get("viewport_width"),
        "viewport_height": dados_json.get("viewport_height"),
        "2dPlayerTracking": {
            "people": []
        }
    }

    # Identifica as coordenadas dos lutadores a partir do JSON
    for pessoa in dados_json["2dPlayerTracking"]["people"]:
        x1, y1, x2, y2 = int(pessoa["left"]), int(pessoa["top"]), int(pessoa["right"]), int(pessoa["bottom"])
        
        # Extrai a área do jogador na imagem
        roi = img[y1:y2, x1:x2]
        
        # Calcular cor média na área para decidir o time
        cor_media = np.mean(roi, axis=(0, 1))  # Média da cor no formato BGR

        # Classificação com base na cor (exemplo: Times Vermelho e Azul)
        if cor_media[2] > cor_media[0]:  # Mais vermelho que azul
            time = "Time 1"
        else:
            time = "Time 0"
        
        # Adiciona as coordenadas e classificação ao novo JSON
        classificacoes["2dPlayerTracking"]["people"].append({
            "left": x1,
            "top": y1,
            "right": x2,
            "bottom": y2,
            "team": time
        })

    # Salvar o JSON com a classificação na pasta de saída
    with open(output_json_path, 'w') as outfile:
        json.dump(classificacoes, outfile, indent=4)

# Processa todas as imagens e JSONs correspondentes
for nome_imagem in os.listdir(pasta_imagens):
    if nome_imagem.endswith((".jpg", ".png")):
        img_path = os.path.join(pasta_imagens, nome_imagem)
        nome_json = os.path.splitext(nome_imagem)[0] + '.json'
        json_path = os.path.join(pasta_json, nome_json)
        output_json_path = os.path.join(pasta_output, nome_json)
        
        # Verifica se o JSON correspondente existe
        if os.path.exists(json_path):
            classificar_lutadores(img_path, json_path, output_json_path)
            print(f"Arquivo JSON classificado salvo em: {output_json_path}")
```

```python
#trasforma em txt
import os
import json

# Caminhos para as pastas de entrada e saída
pasta_json = 'D:/Projeto/hackfot/final-boxe/dataset/train/saidas-final-json'  # Local dos arquivos JSON classificados
pasta_output_txt = 'D:/Projeto/hackfot/final-boxe/dataset/train/saidas-finais-txt'  # Local de saída para os arquivos .txt

# Cria a pasta de saída, se não existir
os.makedirs(pasta_output_txt, exist_ok=True)

# Função para converter coordenadas de boxes de JSON para TXT
def converter_json_para_txt(json_path, output_txt_path, img_width, img_height):
    with open(json_path, 'r') as f:
        dados_json = json.load(f)

    linhas_txt = []

    # Processa cada pessoa no arquivo JSON
    for pessoa in dados_json["2dPlayerTracking"]["people"]:
        # Obter as coordenadas e a classe do time
        x1, y1, x2, y2 = int(pessoa["left"]), int(pessoa["top"]), int(pessoa["right"]), int(pessoa["bottom"])
        team = 1 if pessoa["team"] == "Time 1" else 0

        # Calcular centro e dimensões da box, normalizados
        center_x = ((x1 + x2) / 2) / img_width
        center_y = ((y1 + y2) / 2) / img_height
        width = (x2 - x1) / img_width
        height = (y2 - y1) / img_height

        # Montar a linha no formato especificado
        linha = f"{team} {center_x} {center_y} {width} {height}"
        linhas_txt.append(linha)

    # Salvar o arquivo TXT com as coordenadas normalizadas
    with open(output_txt_path, 'w') as f:
        f.write("\n".join(linhas_txt))

# Processa todos os arquivos JSON na pasta de entrada
for nome_json in os.listdir(pasta_json):
    if nome_json.endswith(".json"):
        json_path = os.path.join(pasta_json, nome_json)
        
        # Define o nome e o caminho para o arquivo de saída .txt
        nome_txt = os.path.splitext(nome_json)[0] + '.txt'
        output_txt_path = os.path.join(pasta_output_txt, nome_txt)
        
        # Definir a largura e altura da imagem (substitua conforme necessário)
        img_width, img_height = 1280, 720  # Ajuste conforme as dimensões reais das imagens

        # Converte o JSON para TXT
        converter_json_para_txt(json_path, output_txt_path, img_width, img_height)
        print(f"Arquivo TXT salvo em: {output_txt_path}")
```

```python
import os
import json

# Caminhos para pastas de entrada e saída
pasta_json_entrada = 'D:/Projeto/hackfot/final-boxe/SAIDA FINAL'
pasta_saida = 'D:/Projeto/hackfot/final-boxe/SAIDA FINAL CONVERTIDA YOLO'

# Criar a pasta de saída, se não existir
os.makedirs(pasta_saida, exist_ok=True)

# Função para converter coordenadas para o formato YOLO
def converter_para_yolo(width, height, left, top, right, bottom):
    x_center = (left + right) / 2.0 / width
    y_center = (top + bottom) / 2.0 / height
    box_width = (right - left) / width
    box_height = (bottom - top) / height
    return x_center, y_center, box_width, box_height

# Processar cada arquivo JSON na pasta de entrada
for arquivo_json in os.listdir(pasta_json_entrada):
    if arquivo_json.endswith(".json"):
        caminho_json = os.path.join(pasta_json_entrada, arquivo_json)
        nome_txt = os.path.splitext(arquivo_json)[0] + '.txt'
        caminho_txt = os.path.join(pasta_saida, nome_txt)
        
        # Abrir o arquivo JSON
        with open(caminho_json, 'r') as f:
            dados = json.load(f)
        
        # Presumindo que a resolução da imagem seja conhecida (substitua pelos valores reais)
        largura_imagem = 1920  # substitua com a largura real da imagem
        altura_imagem = 1080   # substitua com a altura real da imagem

        # Preparar conteúdo para o arquivo de saída
        conteudo_txt = []
        
        for pessoa in dados.get("2dPlayerTracking", {}).get("people", []):
            left = pessoa["left"]
            top = pessoa["top"]
            right = pessoa["right"]
            bottom = pessoa["bottom"]
            team = pessoa["team"]
            
            # Converter classe do time em um número (ex: "Time 1" -> 1, "Time 2" -> 2)
            classe = int(team.split()[-1])

            # Converter coordenadas para o formato YOLO
            x_center, y_center, box_width, box_height = converter_para_yolo(
                largura_imagem, altura_imagem, left, top, right, bottom
            )
            
            # Adicionar linha ao conteúdo do arquivo .txt
            conteudo_txt.append(f"{classe} {x_center} {y_center} {box_width} {box_height}")
        
        # Escrever no arquivo .txt
        with open(caminho_txt, 'w') as f:
            f.write("\n".join(conteudo_txt))
```
